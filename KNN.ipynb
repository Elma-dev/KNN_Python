#<center>**Import Data and somme library**<center/>

```python
#Import anything Here:>
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
```


#<center>**Distance Euclidienne**<center/>
\begin{align}
d({a,b})&=\sqrt{{\sum_{k=0}^n ({x}_a-{x}_b})^2 +({y}_a-{y}_b)^2}
\end{align}

#<center>**KNN Implementation**<center/>

```python
class KNearestNeighbor:
  def __init__(self,K,X_train,Y_train):
    self.K=K
    self.X_train=X_train
    self.Y_train=Y_train
  #----------------get The Dis Euc Between two Point-----------------------#
  def euclidDist(self,row_te,Y):
    e_l=[]
    for  index,row_tr in enumerate(Y):
      s=(row_te - row_tr)**2
      v_euq=np.sqrt(s.sum())
      e_l.append([v_euq,index])
    return e_l
  #----------------get The class Most repeat after calculate the dist------#
  def max_repeat(self,l):
    k=dict()
    for i in l:
      if k.get(i)==None:
        k[i]=1
      else:
        k[i]+=1
    max,m=k[l[0]],l[0]
    for key in k.keys():
      if k[key]>max:
        m=key
    return m
  #----------------test Or Predict-----------------------#
  def test(self,X_test):
    results=[]
    for row in X_test:
      arr_euq=self.euclidDist(row,self.X_train)
      arr_euq=sorted(arr_euq,key=lambda x:x[0])
      arr_euq=np.array(arr_euq)
      arr_euq=arr_euq[:self.K]
      k_predict=[]
      for row in arr_euq:
          k_predict.append(self.Y_train[int(row[1])])
      results.append(self.max_repeat(k_predict))
    return results
  #-----------------Calculate the accurancy----------------#
  def mean_error(self,X_test,Y_test):
    return ((self.test(X_test)==Y_test).sum())/len(Y_test)
```

#<center>**Test :**<center/>
#<center>**in this section we test our algorithm with iris data, we try to predict the type of any flower in our data test.**<center/>

```python
#Split data 
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 42, test_size = 0.2)
#Create KNN Object and gave The K and Data_Train and Data_train
KNN=KNearestNeighbor(10,X_train,y_train)
#Call test method to fit the result  
print(f'result : {KNN.test(y_test)}')
#Calculate The Accuracy
print(f'accurancy : {KNN.mean_error(X_test,y_test)*100}%')
```

#<center>**How I Get The Best K :**</center>
**it's not a magic method :) we just change k after every iteration and plot the accuracy and k and that's the key to how to get the best k.**

```python
K=[]
error=[]
# 
for k in range(1,20):
  KNN=KNearestNeighbor(k,X_train,y_train)
  K.append(k)
  error.append(KNN.mean_error(X_test,y_test))

#plot k--> accurancy
plt.plot(K,error,c="blue",linestyle="--",marker="o",markerfacecolor="red")
plt.xlabel("K")
plt.ylabel("Score")
plt.title("Best K/Score")
plt.show()
```
